```python
from langchain_google_genai import ChatGoogleGenerativeAI
from collections import defaultdict
from PIL import Image
from IPython.display import Image, display
```

```python
annotated_dict = defaultdict(str)
```

```python
import os
import pathlib
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
import google.generativeai as genai
from pathlib import Path
from langchain_core.messages import HumanMessage
os.environ["GOOGLE_API_KEY"] = ""

llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")
```

```python
image_ls = ["data/glioma/Te-gl_0010.jpg", "data/glioma/Te-gl_0011.jpg", 
                    "data/glioma/Te-gl_0012.jpg","data/glioma/Te-gl_0013.jpg", 
                    "data/glioma/Te-gl_0014.jpg","data/glioma/Te-gl_0015.jpg", 
                    "data/glioma/Te-gl_0016.jpg","data/glioma/Te-gl_0017.jpg", 
            ] 

gemini_pro_vision_model = genai.GenerativeModel("gemini-pro-vision")
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")

for i in image_ls:
    message = HumanMessage(
        content=[
            {
                "type": "text",
                "text": """You are tasked to help Radiologists and other doctors create a report of these x-ray images comprising of T2 MRI images. 
    Describe the following image: {image_url} to create a medical report for each file name with a tag. """,
            },
            {"type": "image_url", "image_url": i},
        ]
    )
    response = llm.invoke([message])
    annotated_dict[i] = response
```

```python
annotated_dict = dict(annotated_dict)
```

```python
counter = 0
for k,v in annotated_dict.items():
    print(f'MRI Scan {counter+1}', '\n')
    display(Image(image_ls[counter]))
    print('\n', 'File name', k, 'Description', v.content)
    print('\n\n')
    counter += 1
```
