```python
import os
import pathlib
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
import google.generativeai as genai
from pathlib import Path
from langchain_core.messages import HumanMessage
os.environ["GOOGLE_API_KEY"] = ""

llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")
```

```python

local_image_path = ["Testing/glioma/Te-gl_0010.jpg", "Testing/glioma/Te-gl_0011.jpg", 
                    "Testing/glioma/Te-gl_0012.jpg","Testing/glioma/Te-gl_0013.jpg"] 

# Convert your local image to base64
image_ls = [Image.open(x) for x in local_image_path]

# Use the Gemini Pro Vision model to process the base64 image
gemini_pro_vision_model = genai.GenerativeModel("gemini-pro-vision")
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")

message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": """You are tasked to help Radiologists and other doctors create a report of these x-ray images comprising of T2 MRI images. 
Describe the following image: {image_url} to create a medical report for each file name with a tag. """,
        },
        {"type": "image_url", "image_url": image_ls[0]},
    ]
)
response = llm.invoke([message])
```
